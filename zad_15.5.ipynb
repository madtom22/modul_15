```python
import pandas as pd

prices = [
	(1, 2.12),
	(2, 2.56),
	(3, 3.10),
	(4, 3.16),
	(5, 3.58),
	(6, 5.12),
	(7, 5.16),
	(8, 5.20),
	(9, 4.12),
	(10, 4.10),
	(11, 3.65),
	(12, 4.25),
]

prices_df = pd.DataFrame(prices, columns = ["month","pricePLN"])
```

```python
prices_df = prices_df.set_index("month")
```

```python
prices_df
```

```python
prices_in_USD = prices_df["pricePLN"]/ 4
prices_df["priceUSD"] = prices_in_USD
prices_df
```

```python
import matplotlib.pyplot as plt
import matplotlib as mpl
from cycler import cycler
mpl.rcParams['lines.linewidth'] = 2
mpl.rcParams['lines.linestyle'] = '--'
plt.style.use('fivethirtyeight')
mpl.rcParams['axes.prop_cycle'] = cycler(color='r')
plt.rcParams['axes.facecolor'] = 'white'
plt.title("Price of goods(USD)")
plt.xlabel("months")
plt.plot(prices_df.index, prices_df["priceUSD"])
```
